```python
def remove_outlier(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3-q1
    
    lower_fence = q1-1.5*iqr
    upper_fence = q3 + 1.5*iqr
    
    filter_data = data[~((data<lower_fence) | (data>upper_fence)).any(axis = 1)]
    return filter_data
```

```python
remove=remove_outlier(num_cols)
```

```python
plt.figure(figsize=(15,10))
for i,cols in enumerate(remove.columns):
    plt.subplot(6,6,i+1)
    sns.boxplot(x=remove[cols])
    plt.title(f"{cols}")
plt.tight_layout()
plt.show()
```
